```python
import pandas as pd

# Creaate path to the CSV file

csv_file_path = 'Data/Raw/Heart_Disease_Prediction.csv'

# Load the CSV file into a DataFrame of pandas
try:
    df = pd.read_csv(csv_file_path)
    print("CSV file loaded successfully.")
    print(df.head())
except FileNotFoundError:
    print(f"the file at {csv_file_path} was not found")
```

Now we can see the headers on the dataset

```python
encabezados = df.columns
print("Column headers in the DataFrame:")
for encabezado in encabezados:
    print(encabezado)
```

Now we can start with the Exploration:

```python
# Display basic information about the DataFrame
df.info()
```

```python
# Display summary satistics of the DataFrame
df.describe(include='all')
```

```python
# Drop any null values from the DataFrame
df_cleaned = df.dropna()
print("DataFrame after dropping null values:")
print(df_cleaned.info())
```

We can see that itÂ´s not presence of NaN Data, so, we can continue with the transformation of the data

```python
# Display the  comparison between the age and heart disease
import matplotlib.pyplot as plt
gender = df['Sex']
age = df['Age']

#Now for a better visualization, we need to group the ages by years like 30s, 40s, 50s, etc.
heart_disease = df['Heart Disease']
plt.subplot2grid((1,1), (0,0))
plt.bar(age,heart_disease, color='blue', label='Heart Disease by Age')
plt.xlabel('Age')
plt.show()
```

```python
# 1. Aplicar mapeo (lo que ya hiciste)
mapping = {'Presence': 1, 'Absence': 0}
df['HeartDisease'] = df['HeartDisease'].replace(mapping)

# 2. ð¨ ELIMINAR las filas donde la variable objetivo es nula
# Antes de eliminar, revisa cuÃ¡ntas filas pierdes:
print(f"Filas antes de eliminar: {len(df)}")
df.dropna(subset=['HeartDisease'], inplace=True)
print(f"Filas despuÃ©s de eliminar: {len(df)}")

# 3. Convertir a entero (ahora que no hay NaN)
df['HeartDisease'] = df['HeartDisease'].astype(int)

# Verifica:
print(df['HeartDisease'].unique()) # DeberÃ­a mostrar solo [0 1]
```
