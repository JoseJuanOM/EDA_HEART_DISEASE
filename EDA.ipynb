```python
import pandas as pd

# Creaate path to the CSV file

csv_file_path = 'Data/Raw/Heart_Disease_Prediction.csv'

# Load the CSV file into a DataFrame of pandas
try:
    df = pd.read_csv(csv_file_path)
    print("CSV file loaded successfully.")
    print(df.head())
except FileNotFoundError:
    print(f"the file at {csv_file_path} was not found")
```

Now we can see the headers on the dataset

```python
encabezados = df.columns
print("Column headers in the DataFrame:")
for encabezado in encabezados:
    print(encabezado)
```
